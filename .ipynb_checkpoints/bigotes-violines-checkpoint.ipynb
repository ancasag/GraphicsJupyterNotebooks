```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("kfold-best.csv")
```

```python
methods = df['method']
```

```python
data = [np.array(x*100,dtype=float) for x in df.values[:,1:]]
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(16,16))
axes.violinplot(data,showmeans=False,showmedians=True)
axes.yaxis.grid(True)
axes.set_ylabel("Accuracy")
plt.xticks(range(1,7),methods,rotation='vertical')
plt.savefig("violin.png")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(16,16))
axes.boxplot(data)
axes.yaxis.grid(True)
axes.set_ylabel("Accuracy")
plt.xticks(range(1,7),methods,rotation='vertical')
plt.savefig("boxplot.png")
```
