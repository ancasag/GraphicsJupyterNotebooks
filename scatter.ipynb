```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib
matplotlib.rcParams.update({'font.size': 30})
```

```python
df = pd.read_csv("data/data.csv",sep=";")
```

```python
mlMethods = df.columns.values.tolist()[1:]
```

```python
featureMethods = df.values[:,0].tolist()
```

```python
mlMethods
```

```python
x = np.array(df.values[:,1:].flatten(), dtype=float)
```

```python
y=np.array([np.repeat(i+1,len(mlMethods)) for i,_ in enumerate(featureMethods)]).flatten()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(24,24))
colors=['red','green','blue','purple','pink','orange']
col = []
for _ in featureMethods:
    col.append(colors)
col = np.array(col).flatten()
axes.yaxis.grid(True)
axes.set_xlabel("Accuracy")
axes.xaxis.grid(True)
plt.scatter(x,y,c=col,s=200,label=mlMethods)
plt.yticks(range(1,len(featureMethods)+1),featureMethods)
plt.xlim(80,97.5)
recs=[]
for i in range(0,len(colors)):
    recs.append(mpatches.Circle((0,0),radius=5,fc=colors[i]))
plt.legend(recs,mlMethods,loc=4)
plt.savefig("images/scatter-80-100.png")
```
